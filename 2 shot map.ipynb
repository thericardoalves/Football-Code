```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch, VerticalPitch
```

```python
data = pd.read_csv("/mckayjohns/shotmaps.csv")
```

```python
#The statsbomb pitch from mplsoccer

pitch = VerticalPitch(pitch_type='statsbomb', 
              pitch_color='#22312b', line_color='#c7d5cc',half=True, corner_arcs=True)

fig, ax = pitch.draw()

#invert axis
#plt.gca().invert_yaxis()

#Plotting all shots
plt.scatter(data["x"],data["y"],c="red",s=30,alpha=.5)

#Title
plt.title("Football Shot Map Example", fontsize= 18, c="red")

plt.savefig('shotmaptest.png',dpi=300,bbox_inches = 'tight',facecolor='#22312b')
```
